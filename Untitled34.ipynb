```python
# Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations.
# It is often employed in data analysis, scientific research, and data visualization tasks.
# Matplotlib provides a flexible and extensive toolkit for creating various types of plots and charts, making it a powerful tool for visualizing data and results.

# Line Plot: Line plots are used to visualize data points as connected line segments. They are particularly useful for showing trends, patterns, and changes over time or across different categories.
# Scatter Plot: Scatter plots display individual data points as markers on a two-dimensional plane. They are effective for visualizing relationships, correlations, and distributions of data points, especially when comparing two continuous variables.
# Bar Chart: Bar charts use rectangular bars to represent categorical data. They are often employed to compare the values of different categories or groups, making it easy to see relative differences.
# Histogram: Histograms are used to visualize the distribution of a single variable by dividing the data into bins or intervals and counting the number of data points in each bin. They provide insights into the frequency distribution of data.
# Pie Chart: Pie charts are circular graphs divided into slices, with each slice representing a category's proportion within a whole. They are useful for showing the composition or relative percentages of different categories in a dataset.
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
```

```python
# The subplot() function in Matplotlib is used to create multiple subplots (or small plots) within a single figure.
# It allows you to arrange and display multiple plots or charts in a grid-like structure, making it easier to compare and visualize different aspects of your data in the same figure.
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(10,8))
```

```python
plt.subplot(2,2,1)
plt.plot(x1,y1)

plt.subplot(2,2,2)
plt.plot(x2,y2)

plt.subplot(2,2,3)
plt.plot(x3,y3)

plt.subplot(2,2,4)
plt.plot(x4,y4)

plt.tight_layout()
plt.show()
```

```python
# A bar plot, also known as a bar chart or bar graph, is a common type of data visualization used to represent categorical data.
# It displays data using rectangular bars, where the length or height of each bar is proportional to the value it represents. Bar plots are used to visually compare and display the distribution or relationships between different categories or groups of data.
# They are particularly useful for showing discrete data and making comparisons between individual categories.
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

```python
# A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's distribution.
# It displays the distribution of a dataset using five summary statistics: the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.
# Box plots are used to visualize the spread and skewness of the data, identify outliers, and compare the distributions of different datasets.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.show()
```

```python
plt.boxplot(box2)
plt.show()
```
