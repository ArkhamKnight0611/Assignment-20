```python
import plotly.express as px
```

```python
df=px.data.tips()
fig=px.box(df,x="day", y="total_bill", color="smoker")
fig.show()
```

```python
fig = px.box(df, x='day', y='tip')
fig.show()
```

```python
px.box(df, y="total_bill")
```

```python
df = px.data.tips()
fig = px.histogram(df, x="sex", y="total_bill", color="day", pattern_shape="smoker")
fig.show()
```

```python
# A distplot, short for "distribution plot," is a data visualization that combines a histogram with a kernel density estimate (KDE) plot.
# It provides a way to visualize the distribution of a univariate dataset, showing the frequency of data points within different bins as well as an estimated smooth probability density curve.
```
