```python
# Seaborn is a powerful data visualization library in Python that provides various types of plots for exploring and visualizing data.

# Scatter Plot (sns.scatterplot):
#Use: Scatter plots are used to visualize the relationship between two continuous variables. Each data point is represented as a dot, and you can use them to identify patterns, correlations, or outliers in your data.

#Line Plot (sns.lineplot):
#Use: Line plots are used to visualize trends in data over a continuous variable, typically time. They are great for showing how a variable changes over a specific range and can be used for time series data.

#Bar Plot (sns.barplot):
#Use: Bar plots are used to display the distribution of a categorical variable. They show the relationship between a categorical variable and a continuous variable, typically by displaying the mean or another statistic of the continuous variable for each category.

#Histogram (sns.histplot):
#Use: Histograms are used to visualize the distribution of a single continuous variable. They divide the data into bins and display the frequency or density of data points within each bin. Histograms help you understand the data's underlying distribution.

#Box Plot (sns.boxplot):
#Use: Box plots are used to display the summary statistics (median, quartiles, and potential outliers) of a continuous variable across different categories or groups. They are useful for detecting variations and identifying outliers within each category.
```

```python
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot_table(index="month",columns="year",values="passengers")
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True,fmt='d')
plt.title("Passenger Count by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
