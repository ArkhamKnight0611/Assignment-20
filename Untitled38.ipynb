```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure ,output_file , show
from bokeh.sampledata.iris import flowers
```

```python
x=[1,3,5,7,9,12]
y=[5,10,12,14,19,22]
p = figure(title="line_plot")
p.line(x,y,line_width=2)
show(p)
```

```python
# Glyphs in Bokeh are visual elements used to represent data points or shapes on a plot.
# Glyphs can take various forms, such as circles, lines, rectangles, and more, and they allow you to display data in a visually meaningful way.
# Bokeh provides a wide range of glyph types to create different types of plots, and you can customize these glyphs with various attributes like color, size, and transparency to enhance data visualization.
```

```python
x=[1,3,5,7,9,12]
y=[5,10,12,14,19,22]
p = figure(title="line_plot")
p.circle(x,y,size=10, color="blue", legend_label="Circles")
show(p)
```

```python
x=[1,3,5,7,9,12]
y=[5,10,12,14,19,22]
p = figure(title="line_plot")
p.square(x,y,size=10, color="blue", legend_label="Circles")
show(p)
```
